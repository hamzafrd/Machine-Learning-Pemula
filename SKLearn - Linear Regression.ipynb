```python
#memprediksi harga rumah berdasarkan jumlah kamar.

#membuat dummy dengan library numpy
import numpy as np

#data jumlah kamar
kamar = np.array([1,1,2,2,3,4,4,5,5,5])

#harga rumah (dalam dolar)
harga_rumah = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])

#menampilkan data dengan scatter plot
import matplotlib.pyplot as plt
%matplotlib inline

# plt.scatter(kamar, harga_rumah)

#melatih model
from sklearn.linear_model import LinearRegression

#latih dengan fit()
kamar = kamar.reshape(-1,1)
linreg = LinearRegression()
linreg.fit(kamar, harga_rumah)

#menampilkan plot- hubungan antar jumlah kamar dan harga rumah
plt.scatter(kamar, harga_rumah) #titik bulet
plt.plot(kamar, linreg.predict(kamar)) #garis prediksi
```

```python
#import csv dengan pandas
import pandas as pd
iklan = pd.read_csv('Social_Network_Ads.csv')
# iklan.head()
# iklan.info()

# delete id
data = iklan.drop(columns=['User ID'])

#jalankan one-hot encoding dengan pd.dummies (merubah data huruf menjadi 0 1)
data = pd.get_dummies(data)
data

# seperti biasa, abis di drop di pisahkan atribut dan label
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']
X = data[predictions]
y = data['Purchased'] 

# normalisasi
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)

scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data,columns=X.columns)
scaled_data.head()

#Data set (test dan train)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
 
# latih model dengan fungsi fit
from sklearn import linear_model
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)

# uji akurasi model
model.score(X_test, y_test)
```
